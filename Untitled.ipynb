```python
import googlemaps
import pandas as pd
import gspread
import time
from oauth2client.service_account import ServiceAccountCredentials
from math import radians, cos, sin, asin, sqrt

# 例: mareの緯度経度
api_key = "AIzaSyCCPxWFw3oKoFXaMZnjBWXp99PkuXIsZ4k"
place_types = [
     'bar',
     'restaurant',
     'cafe',
     'night_club'
     ]

fields = [
        'name',
        'rating',
        'business_status',
        'user_ratings_total',
        'geometry',
        'opening_hours',
        'price_level',
        'formatted_address',
        'formatted_phone_number',
        'website',
        'reviews'
    ]


# RenSa（基準点）　
base_lat = 26.221361068312497
base_lng = 127.67789905821657


points = {
    "point_number_1": {
        "latitude": 26.221361068312497,
        "longitude": 127.67789905821657,
        "radius": 50
    },
    "point_number_2": {
        "latitude": 26.221361068312497,
        "longitude": 127.6784002999493,
        "radius": 50
    },
    "point_number_3": {
        "latitude": 26.221750485990945,
        "longitude": 127.67814967908294,
        "radius": 50
    },
    "point_number_4": {
        "latitude": 26.221750485990945,
        "longitude": 127.6776484373502,
        "radius": 50
    },
    "point_number_5": {
        "latitude": 26.221361068312497,
        "longitude": 127.67739781648383,
        "radius": 50
    },
    "point_number_6": {
        "latitude": 26.22097165063405,
        "longitude": 127.6776484373502,
        "radius": 50
    },
    "point_number_7": {
        "latitude": 26.22097165063405,
        "longitude": 127.67814967908294,
        "radius": 50
    },
    "point_number_8": {
        "latitude": 26.221361068312497,
        "longitude": 127.67890154168204,
        "radius": 50
    },
    "point_number_9": {
        "latitude": 26.221810729115457,
        "longitude": 127.67876723436454,
        "radius": 50
    },
    "point_number_10": {
        "latitude": 26.222139903669394,
        "longitude": 127.6784002999493,
        "radius": 50
    },
    "point_number_11": {
        "latitude": 26.222260389918414,
        "longitude": 127.67789905821657,
        "radius": 50
    },
    "point_number_12": {
        "latitude": 26.222139903669394,
        "longitude": 127.67739781648383,
        "radius": 50
    },
    "point_number_13": {
        "latitude": 26.221810729115457,
        "longitude": 127.67703088206859,
        "radius": 50
    },
    "point_number_14": {
        "latitude": 26.221361068312497,
        "longitude": 127.6768965747511,
        "radius": 50
    },
    "point_number_15": {
        "latitude": 26.220911407509536,
        "longitude": 127.67703088206859,
        "radius": 50
    },
    "point_number_16": {
        "latitude": 26.2205822329556,
        "longitude": 127.67739781648383,
        "radius": 50
    },
    "point_number_17": {
        "latitude": 26.22046174670658,
        "longitude": 127.67789905821657,
        "radius": 50
    },
    "point_number_18": {
        "latitude": 26.2205822329556,
        "longitude": 127.6784002999493,
        "radius": 50
    },
    "point_number_19": {
        "latitude": 26.220911407509536,
        "longitude": 127.67876723436454,
        "radius": 50
    },
    "point_number_20": {
        "latitude": 26.221361068312497,
        "longitude": 127.67940278341477,
        "radius": 50
    },
    "point_number_21": {
        "latitude": 26.221822447469325,
        "longitude": 127.679312097689,
        "radius": 50
    },
    "point_number_22": {
        "latitude": 26.22222817749061,
        "longitude": 127.67905097854863,
        "radius": 50
    },
    "point_number_23": {
        "latitude": 26.222529321347842,
        "longitude": 127.67865092081567,
        "radius": 50
    },
    "point_number_24": {
        "latitude": 26.22268955664744,
        "longitude": 127.67816017735694,
        "radius": 50
    },
    "point_number_25": {
        "latitude": 26.22268955664744,
        "longitude": 127.67763793907619,
        "radius": 50
    },
    "point_number_26": {
        "latitude": 26.222529321347842,
        "longitude": 127.67714719561747,
        "radius": 50
    },
    "point_number_27": {
        "latitude": 26.22222817749061,
        "longitude": 127.6767471378845,
        "radius": 50
    },
    "point_number_28": {
        "latitude": 26.221822447469325,
        "longitude": 127.67648601874413,
        "radius": 50
    },
    "point_number_29": {
        "latitude": 26.221361068312497,
        "longitude": 127.67639533301836,
        "radius": 50
    },
    "point_number_30": {
        "latitude": 26.22089968915567,
        "longitude": 127.67648601874413,
        "radius": 50
    },
    "point_number_31": {
        "latitude": 26.220493959134384,
        "longitude": 127.6767471378845,
        "radius": 50
    },
    "point_number_32": {
        "latitude": 26.22019281527715,
        "longitude": 127.67714719561747,
        "radius": 50
    },
    "point_number_33": {
        "latitude": 26.220032579977556,
        "longitude": 127.67763793907619,
        "radius": 50
    },
    "point_number_34": {
        "latitude": 26.220032579977556,
        "longitude": 127.67816017735694,
        "radius": 50
    },
    "point_number_35": {
        "latitude": 26.22019281527715,
        "longitude": 127.67865092081567,
        "radius": 50
    },
    "point_number_36": {
        "latitude": 26.220493959134384,
        "longitude": 127.67905097854863,
        "radius": 50
    },
    "point_number_37": {
        "latitude": 26.22089968915567,
        "longitude": 127.679312097689,
        "radius": 50
    },
    "point_number_38": {
        "latitude": 26.221361068312497,
        "longitude": 127.67990402514751,
        "radius": 50
    },
    "point_number_39": {
        "latitude": 26.2218083726899,
        "longitude": 127.6798410354245,
        "radius": 50
    },
    "point_number_40": {
        "latitude": 26.222227571288197,
        "longitude": 127.67965602413142,
        "radius": 50
    },
    "point_number_41": {
        "latitude": 26.22259232431776,
        "longitude": 127.67936061620824,
        "radius": 50
    },
    "point_number_42": {
        "latitude": 26.222879713004264,
        "longitude": 127.67897337322121,
        "radius": 50
    },
    "point_number_43": {
        "latitude": 26.223071679659604,
        "longitude": 127.67851862707138,
        "radius": 50
    },
    "point_number_44": {
        "latitude": 26.223156162312815,
        "longitude": 127.6780249511318,
        "radius": 50
    },
    "point_number_45": {
        "latitude": 26.223127852608094,
        "longitude": 127.67752336487735,
        "radius": 50
    },
    "point_number_46": {
        "latitude": 26.222988529348303,
        "longitude": 127.6770453848171,
        "radius": 50
    },
    "point_number_47": {
        "latitude": 26.222746946726247,
        "longitude": 127.67662104419614,
        "radius": 50
    },
    "point_number_48": {
        "latitude": 26.22241828426655,
        "longitude": 127.67627700589628,
        "radius": 50
    },
    "point_number_49": {
        "latitude": 26.222023193040293,
        "longitude": 127.6760348871092,
        "radius": 50
    },
    "point_number_50": {
        "latitude": 26.221586498082264,
        "longitude": 127.67590990104873,
        "radius": 50
    },
    "point_number_51": {
        "latitude": 26.22113563854273,
        "longitude": 127.67590990104873,
        "radius": 50
    },
    "point_number_52": {
        "latitude": 26.2206989435847,
        "longitude": 127.6760348871092,
        "radius": 50
    },
    "point_number_53": {
        "latitude": 26.220303852358445,
        "longitude": 127.67627700589628,
        "radius": 50
    },
    "point_number_54": {
        "latitude": 26.219975189898747,
        "longitude": 127.67662104419614,
        "radius": 50
    },
    "point_number_55": {
        "latitude": 26.21973360727669,
        "longitude": 127.6770453848171,
        "radius": 50
    },
    "point_number_56": {
        "latitude": 26.2195942840169,
        "longitude": 127.67752336487735,
        "radius": 50
    },
    "point_number_57": {
        "latitude": 26.21956597431218,
        "longitude": 127.6780249511318,
        "radius": 50
    },
    "point_number_58": {
        "latitude": 26.21965045696539,
        "longitude": 127.67851862707138,
        "radius": 50
    },
    "point_number_59": {
        "latitude": 26.21984242362073,
        "longitude": 127.67897337322121,
        "radius": 50
    },
    "point_number_60": {
        "latitude": 26.220129812307235,
        "longitude": 127.67936061620824,
        "radius": 50
    },
    "point_number_61": {
        "latitude": 26.220494565336796,
        "longitude": 127.67965602413142,
        "radius": 50
    },
    "point_number_62": {
        "latitude": 26.220913763935094,
        "longitude": 127.6798410354245,
        "radius": 50
    },
    "point_number_63": {
        "latitude": 26.221361068312497,
        "longitude": 127.68040526688024,
        "radius": 50
    },
    "point_number_64": {
        "latitude": 26.221813648583385,
        "longitude": 127.68035396464167,
        "radius": 50
    },
    "point_number_65": {
        "latitude": 26.222247700164807,
        "longitude": 127.6802021582456,
        "radius": 50
    },
    "point_number_66": {
        "latitude": 26.22264545293402,
        "longitude": 127.67995606266375,
        "radius": 50
    },
    "point_number_67": {
        "latitude": 26.22299062284592,
        "longitude": 127.67962575307814,
        "radius": 50
    },
    "point_number_68": {
        "latitude": 26.223269078603757,
        "longitude": 127.67922475240202,
        "radius": 50
    },
    "point_number_69": {
        "latitude": 26.22346942019609,
        "longitude": 127.67876947765019,
        "radius": 50
    },
    "point_number_70": {
        "latitude": 26.223583445614583,
        "longitude": 127.67827856782448,
        "radius": 50
    },
    "point_number_71": {
        "latitude": 26.223606486645213,
        "longitude": 127.67777212083081,
        "radius": 50
    },
    "point_number_72": {
        "latitude": 26.223537599985477,
        "longitude": 127.67727087066865,
        "radius": 50
    },
    "point_number_73": {
        "latitude": 26.223379605863318,
        "longitude": 127.6767953385785,
        "radius": 50
    },
    "point_number_74": {
        "latitude": 26.223138972576656,
        "longitude": 127.67636499290002,
        "radius": 50
    },
    "point_number_75": {
        "latitude": 26.22282555168053,
        "longitude": 127.67599745203584,
        "radius": 50
    },
    "point_number_76": {
        "latitude": 26.222452174663243,
        "longitude": 127.67570776315213,
        "radius": 50
    },
    "point_number_77": {
        "latitude": 26.22203412762376,
        "longitude": 127.67550778614583,
        "radius": 50
    },
    "point_number_78": {
        "latitude": 26.22158852545699,
        "longitude": 127.67540570809905,
        "radius": 50
    },
    "point_number_79": {
        "latitude": 26.221133611168003,
        "longitude": 127.67540570809905,
        "radius": 50
    },
    "point_number_80": {
        "latitude": 26.220688009001233,
        "longitude": 127.67550778614583,
        "radius": 50
    },
    "point_number_81": {
        "latitude": 26.22026996196175,
        "longitude": 127.67570776315213,
        "radius": 50
    },
    "point_number_82": {
        "latitude": 26.219896584944465,
        "longitude": 127.67599745203584,
        "radius": 50
    },
    "point_number_83": {
        "latitude": 26.219583164048338,
        "longitude": 127.67636499290002,
        "radius": 50
    },
    "point_number_84": {
        "latitude": 26.219342530761676,
        "longitude": 127.6767953385785,
        "radius": 50
    },
    "point_number_85": {
        "latitude": 26.219184536639517,
        "longitude": 127.67727087066865,
        "radius": 50
    },
    "point_number_86": {
        "latitude": 26.21911564997978,
        "longitude": 127.67777212083081,
        "radius": 50
    },
    "point_number_87": {
        "latitude": 26.21913869101041,
        "longitude": 127.67827856782448,
        "radius": 50
    },
    "point_number_88": {
        "latitude": 26.219252716428905,
        "longitude": 127.67876947765019,
        "radius": 50
    },
    "point_number_89": {
        "latitude": 26.219453058021237,
        "longitude": 127.67922475240202,
        "radius": 50
    },
    "point_number_90": {
        "latitude": 26.219731513779074,
        "longitude": 127.67962575307814,
        "radius": 50
    },
    "point_number_91": {
        "latitude": 26.220076683690973,
        "longitude": 127.67995606266375,
        "radius": 50
    },
    "point_number_92": {
        "latitude": 26.220474436460186,
        "longitude": 127.6802021582456,
        "radius": 50
    },
    "point_number_93": {
        "latitude": 26.22090848804161,
        "longitude": 127.68035396464167,
        "radius": 50
    },
    "point_number_94": {
        "latitude": 26.221361068312497,
        "longitude": 127.68090650861298,
        "radius": 50
    },
    "point_number_95": {
        "latitude": 26.2218170265799,
        "longitude": 127.68086324917685,
        "radius": 50
    },
    "point_number_96": {
        "latitude": 26.222259867758105,
        "longitude": 127.68073471536368,
        "radius": 50
    },
    "point_number_97": {
        "latitude": 26.222676852112702,
        "longitude": 127.68052460485727,
        "radius": 50
    },
    "point_number_98": {
        "latitude": 26.22305598376301,
        "longitude": 127.68023896215432,
        "radius": 50
    },
    "point_number_99": {
        "latitude": 26.223386355781734,
        "longitude": 127.67988600467532,
        "radius": 50
    },
    "point_number_100": {
        "latitude": 26.223658463967396,
        "longitude": 127.67947588636434,
        "radius": 50
    },
    "point_number_101": {
        "latitude": 26.223864480262925,
        "longitude": 127.67902040557847,
        "radius": 50
    },
    "point_number_102": {
        "latitude": 26.22399847795459,
        "longitude": 127.67853266567063,
        "radius": 50
    },
    "point_number_103": {
        "latitude": 26.224056602172773,
        "longitude": 127.67802669802994,
        "radius": 50
    },
    "point_number_104": {
        "latitude": 26.22403718078954,
        "longitude": 127.67751705842421,
        "radius": 50
    },
    "point_number_105": {
        "latitude": 26.22394077252273,
        "longitude": 127.67701840825701,
        "radius": 50
    },
    "point_number_106": {
        "latitude": 26.22377015086265,
        "longitude": 127.67654509278574,
        "radius": 50
    },
    "point_number_107": {
        "latitude": 26.223530224283802,
        "longitude": 127.67611072843476,
        "radius": 50
    },
    "point_number_108": {
        "latitude": 26.223227895037038,
        "longitude": 127.67572781107562,
        "radius": 50
    },
    "point_number_109": {
        "latitude": 26.222871860584327,
        "longitude": 127.67540735654352,
        "radius": 50
    },
    "point_number_110": {
        "latitude": 26.22247236338865,
        "longitude": 127.67515858373193,
        "radius": 50
    },
    "point_number_111": {
        "latitude": 26.22204089625696,
        "longitude": 127.67498864938169,
        "radius": 50
    },
    "point_number_112": {
        "latitude": 26.221589871713068,
        "longitude": 127.67490244219466,
        "radius": 50
    },
    "point_number_113": {
        "latitude": 26.221132264911926,
        "longitude": 127.67490244219466,
        "radius": 50
    },
    "point_number_114": {
        "latitude": 26.220681240368034,
        "longitude": 127.67498864938169,
        "radius": 50
    },
    "point_number_115": {
        "latitude": 26.220249773236343,
        "longitude": 127.67515858373193,
        "radius": 50
    },
    "point_number_116": {
        "latitude": 26.219850276040667,
        "longitude": 127.67540735654352,
        "radius": 50
    },
    "point_number_117": {
        "latitude": 26.219494241587956,
        "longitude": 127.67572781107562,
        "radius": 50
    },
    "point_number_118": {
        "latitude": 26.21919191234119,
        "longitude": 127.67611072843476,
        "radius": 50
    },
    "point_number_119": {
        "latitude": 26.218951985762345,
        "longitude": 127.67654509278574,
        "radius": 50
    },
    "point_number_120": {
        "latitude": 26.218781364102263,
        "longitude": 127.67701840825701,
        "radius": 50
    },
    "point_number_121": {
        "latitude": 26.218684955835453,
        "longitude": 127.67751705842421,
        "radius": 50
    },
    "point_number_122": {
        "latitude": 26.21866553445222,
        "longitude": 127.67802669802994,
        "radius": 50
    },
    "point_number_123": {
        "latitude": 26.218723658670402,
        "longitude": 127.67853266567063,
        "radius": 50
    },
    "point_number_124": {
        "latitude": 26.21885765636207,
        "longitude": 127.67902040557847,
        "radius": 50
    },
    "point_number_125": {
        "latitude": 26.219063672657597,
        "longitude": 127.67947588636434,
        "radius": 50
    },
    "point_number_126": {
        "latitude": 26.21933578084326,
        "longitude": 127.67988600467532,
        "radius": 50
    },
    "point_number_127": {
        "latitude": 26.219666152861983,
        "longitude": 127.68023896215432,
        "radius": 50
    },
    "point_number_128": {
        "latitude": 26.22004528451229,
        "longitude": 127.68052460485727,
        "radius": 50
    },
    "point_number_129": {
        "latitude": 26.22046226886689,
        "longitude": 127.68073471536368,
        "radius": 50
    },
    "point_number_130": {
        "latitude": 26.220905110045095,
        "longitude": 127.68086324917685,
        "radius": 50
    },
    "point_number_131": {
        "latitude": 26.221361068312497,
        "longitude": 127.68140775034571,
        "radius": 50
    },
    "point_number_132": {
        "latitude": 26.221819366278467,
        "longitude": 127.68137035950515,
        "radius": 50
    },
    "point_number_133": {
        "latitude": 26.222267896418963,
        "longitude": 127.68125898390433,
        "radius": 50
    },
    "point_number_134": {
        "latitude": 26.222697099092763,
        "longitude": 127.68107599732082,
        "radius": 50
    },
    "point_number_135": {
        "latitude": 26.22309782659003,
        "longitude": 127.68082529979593,
        "radius": 50
    },
    "point_number_136": {
        "latitude": 26.223461538099887,
        "longitude": 127.68051223451216,
        "radius": 50
    },
    "point_number_137": {
        "latitude": 26.223780481742953,
        "longitude": 127.68014347391254,
        "radius": 50
    },
    "point_number_138": {
        "latitude": 26.224047859789216,
        "longitude": 127.67972687748899,
        "radius": 50
    },
    "point_number_139": {
        "latitude": 26.224257973539828,
        "longitude": 127.67927132427103,
        "radius": 50
    },
    "point_number_140": {
        "latitude": 26.22440634478498,
        "longitude": 127.67878652358456,
        "radius": 50
    },
    "point_number_141": {
        "latitude": 26.224489811249178,
        "longitude": 127.67828280811456,
        "radius": 50
    },
    "point_number_142": {
        "latitude": 26.22450659398965,
        "longitude": 127.67777091368173,
        "radius": 50
    },
    "point_number_143": {
        "latitude": 26.224456335311416,
        "longitude": 127.67726175042718,
        "radius": 50
    },
    "point_number_144": {
        "latitude": 26.22434010639095,
        "longitude": 127.6767661702817,
        "radius": 50
    },
    "point_number_145": {
        "latitude": 26.224160384445902,
        "longitude": 127.67629473567567,
        "radius": 50
    },
    "point_number_146": {
        "latitude": 26.2239209999375,
        "longitude": 127.67585749441928,
        "radius": 50
    },
    "point_number_147": {
        "latitude": 26.22362705493094,
        "longitude": 127.67546376555075,
        "radius": 50
    },
    "point_number_148": {
        "latitude": 26.2232848143537,
        "longitude": 127.67512194071736,
        "radius": 50
    },
    "point_number_149": {
        "latitude": 26.222901572469503,
        "longitude": 127.67483930532198,
        "radius": 50
    },
    "point_number_150": {
        "latitude": 26.222485497413746,
        "longitude": 127.67462188324735,
        "radius": 50
    },
    "point_number_151": {
        "latitude": 26.222045457103874,
        "longitude": 127.67447430846754,
        "radius": 50
    },
    "point_number_152": {
        "latitude": 26.221590830235115,
        "longitude": 127.67439972628273,
        "radius": 50
    },
    "point_number_153": {
        "latitude": 26.22113130638988,
        "longitude": 127.67439972628273,
        "radius": 50
    },
    "point_number_154": {
        "latitude": 26.22067667952112,
        "longitude": 127.67447430846754,
        "radius": 50
    },
    "point_number_155": {
        "latitude": 26.220236639211247,
        "longitude": 127.67462188324735,
        "radius": 50
    },
    "point_number_156": {
        "latitude": 26.21982056415549,
        "longitude": 127.67483930532198,
        "radius": 50
    },
    "point_number_157": {
        "latitude": 26.219437322271293,
        "longitude": 127.67512194071736,
        "radius": 50
    },
    "point_number_158": {
        "latitude": 26.219095081694054,
        "longitude": 127.67546376555075,
        "radius": 50
    },
    "point_number_159": {
        "latitude": 26.218801136687492,
        "longitude": 127.67585749441928,
        "radius": 50
    },
    "point_number_160": {
        "latitude": 26.21856175217909,
        "longitude": 127.67629473567567,
        "radius": 50
    },
    "point_number_161": {
        "latitude": 26.218382030234043,
        "longitude": 127.6767661702817,
        "radius": 50
    },
    "point_number_162": {
        "latitude": 26.218265801313578,
        "longitude": 127.67726175042718,
        "radius": 50
    },
    "point_number_163": {
        "latitude": 26.218215542635345,
        "longitude": 127.67777091368173,
        "radius": 50
    },
    "point_number_164": {
        "latitude": 26.218232325375816,
        "longitude": 127.67828280811456,
        "radius": 50
    },
    "point_number_165": {
        "latitude": 26.218315791840013,
        "longitude": 127.67878652358456,
        "radius": 50
    },
    "point_number_166": {
        "latitude": 26.218464163085166,
        "longitude": 127.67927132427103,
        "radius": 50
    },
    "point_number_167": {
        "latitude": 26.218674276835777,
        "longitude": 127.67972687748899,
        "radius": 50
    },
    "point_number_168": {
        "latitude": 26.21894165488204,
        "longitude": 127.68014347391254,
        "radius": 50
    },
    "point_number_169": {
        "latitude": 26.219260598525107,
        "longitude": 127.68051223451216,
        "radius": 50
    },
    "point_number_170": {
        "latitude": 26.219624310034963,
        "longitude": 127.68082529979593,
        "radius": 50
    },
    "point_number_171": {
        "latitude": 26.22002503753223,
        "longitude": 127.68107599732082,
        "radius": 50
    },
    "point_number_172": {
        "latitude": 26.22045424020603,
        "longitude": 127.68125898390433,
        "radius": 50
    },
    "point_number_173": {
        "latitude": 26.220902770346527,
        "longitude": 127.68137035950515,
        "radius": 50
    },
    "point_number_174": {
        "latitude": 26.221361068312497,
        "longitude": 127.68190899207845,
        "radius": 50
    },
    "point_number_175": {
        "latitude": 26.221811927852034,
        "longitude": 127.68187737255224,
        "radius": 50
    },
    "point_number_176": {
        "latitude": 26.2222556770673,
        "longitude": 127.68178301263242,
        "radius": 50
    },
    "point_number_177": {
        "latitude": 26.22268531776809,
        "longitude": 127.68162740043131,
        "radius": 50
    },
    "point_number_178": {
        "latitude": 26.223094074263898,
        "longitude": 127.68141299004627,
        "radius": 50
    },
    "point_number_179": {
        "latitude": 26.223475500220605,
        "longitude": 127.68114316285715,
        "radius": 50
    },
    "point_number_180": {
        "latitude": 26.22382358032302,
        "longitude": 127.68082217419992,
        "radius": 50
    },
    "point_number_181": {
        "latitude": 26.224132825139996,
        "longitude": 127.68045508625744,
        "radius": 50
    },
    "point_number_182": {
        "latitude": 26.224398357696035,
        "longitude": 127.68004768822586,
        "radius": 50
    },
    "point_number_183": {
        "latitude": 26.224615990384105,
        "longitude": 127.6796064050155,
        "radius": 50
    },
    "point_number_184": {
        "latitude": 26.22478229100671,
        "longitude": 127.67913819592621,
        "radius": 50
    },
    "point_number_185": {
        "latitude": 26.22489463690369,
        "longitude": 127.67865044489501,
        "radius": 50
    },
    "point_number_186": {
        "latitude": 26.224951256313137,
        "longitude": 127.67815084404704,
        "radius": 50
    },
    "point_number_187": {
        "latitude": 26.224951256313137,
        "longitude": 127.6776472723861,
        "radius": 50
    },
    "point_number_188": {
        "latitude": 26.22489463690369,
        "longitude": 127.67714767153812,
        "radius": 50
    },
    "point_number_189": {
        "latitude": 26.22478229100671,
        "longitude": 127.67665992050692,
        "radius": 50
    },
    "point_number_190": {
        "latitude": 26.224615990384105,
        "longitude": 127.67619171141763,
        "radius": 50
    },
    "point_number_191": {
        "latitude": 26.224398357696035,
        "longitude": 127.67575042820728,
        "radius": 50
    },
    "point_number_192": {
        "latitude": 26.224132825139996,
        "longitude": 127.6753430301757,
        "radius": 50
    },
    "point_number_193": {
        "latitude": 26.22382358032302,
        "longitude": 127.67497594223322,
        "radius": 50
    },
    "point_number_194": {
        "latitude": 26.223475500220605,
        "longitude": 127.67465495357598,
        "radius": 50
    },
    "point_number_195": {
        "latitude": 26.223094074263898,
        "longitude": 127.67438512638687,
        "radius": 50
    },
    "point_number_196": {
        "latitude": 26.22268531776809,
        "longitude": 127.67417071600183,
        "radius": 50
    },
    "point_number_197": {
        "latitude": 26.2222556770673,
        "longitude": 127.67401510380071,
        "radius": 50
    },
    "point_number_198": {
        "latitude": 26.221811927852034,
        "longitude": 127.6739207438809,
        "radius": 50
    },
    "point_number_199": {
        "latitude": 26.221361068312497,
        "longitude": 127.67388912435469,
        "radius": 50
    },
    "point_number_200": {
        "latitude": 26.22091020877296,
        "longitude": 127.6739207438809,
        "radius": 50
    },
    "point_number_201": {
        "latitude": 26.220466459557695,
        "longitude": 127.67401510380071,
        "radius": 50
    },
    "point_number_202": {
        "latitude": 26.220036818856904,
        "longitude": 127.67417071600183,
        "radius": 50
    },
    "point_number_203": {
        "latitude": 26.219628062361096,
        "longitude": 127.67438512638687,
        "radius": 50
    },
    "point_number_204": {
        "latitude": 26.21924663640439,
        "longitude": 127.67465495357598,
        "radius": 50
    },
    "point_number_205": {
        "latitude": 26.218898556301973,
        "longitude": 127.67497594223322,
        "radius": 50
    },
    "point_number_206": {
        "latitude": 26.218589311484997,
        "longitude": 127.6753430301757,
        "radius": 50
    },
    "point_number_207": {
        "latitude": 26.21832377892896,
        "longitude": 127.67575042820728,
        "radius": 50
    },
    "point_number_208": {
        "latitude": 26.21810614624089,
        "longitude": 127.67619171141763,
        "radius": 50
    },
    "point_number_209": {
        "latitude": 26.217939845618282,
        "longitude": 127.67665992050692,
        "radius": 50
    },
    "point_number_210": {
        "latitude": 26.217827499721302,
        "longitude": 127.67714767153812,
        "radius": 50
    },
    "point_number_211": {
        "latitude": 26.217770880311857,
        "longitude": 127.6776472723861,
        "radius": 50
    },
    "point_number_212": {
        "latitude": 26.217770880311857,
        "longitude": 127.67815084404704,
        "radius": 50
    },
    "point_number_213": {
        "latitude": 26.217827499721302,
        "longitude": 127.67865044489501,
        "radius": 50
    },
    "point_number_214": {
        "latitude": 26.217939845618282,
        "longitude": 127.67913819592621,
        "radius": 50
    },
    "point_number_215": {
        "latitude": 26.21810614624089,
        "longitude": 127.6796064050155,
        "radius": 50
    },
    "point_number_216": {
        "latitude": 26.21832377892896,
        "longitude": 127.68004768822586,
        "radius": 50
    },
    "point_number_217": {
        "latitude": 26.218589311484997,
        "longitude": 127.68045508625744,
        "radius": 50
    },
    "point_number_218": {
        "latitude": 26.218898556301973,
        "longitude": 127.68082217419992,
        "radius": 50
    },
    "point_number_219": {
        "latitude": 26.21924663640439,
        "longitude": 127.68114316285715,
        "radius": 50
    },
    "point_number_220": {
        "latitude": 26.219628062361096,
        "longitude": 127.68141299004627,
        "radius": 50
    },
    "point_number_221": {
        "latitude": 26.220036818856904,
        "longitude": 127.68162740043131,
        "radius": 50
    },
    "point_number_222": {
        "latitude": 26.220466459557695,
        "longitude": 127.68178301263242,
        "radius": 50
    },
    "point_number_223": {
        "latitude": 26.22091020877296,
        "longitude": 127.68187737255224,
        "radius": 50
    },
    "point_number_224": {
        "latitude": 26.221361068312497,
        "longitude": 127.68241023381118,
        "radius": 50
    },
    "point_number_225": {
        "latitude": 26.221814182638546,
        "longitude": 127.68238186848592,
        "radius": 50
    },
    "point_number_226": {
        "latitude": 26.22226159878904,
        "longitude": 127.68229712922052,
        "radius": 50
    },
    "point_number_227": {
        "latitude": 26.222697690246292,
        "longitude": 127.68215708166042,
        "radius": 50
    },
    "point_number_228": {
        "latitude": 26.2231169729072,
        "longitude": 127.68196348698496,
        "radius": 50
    },
    "point_number_229": {
        "latitude": 26.223514174049033,
        "longitude": 127.6817187797595,
        "radius": 50
    },
    "point_number_230": {
        "latitude": 26.223884298636964,
        "longitude": 127.68142603731938,
        "radius": 50
    },
    "point_number_231": {
        "latitude": 26.224222692139556,
        "longitude": 127.68108894107064,
        "radius": 50
    },
    "point_number_232": {
        "latitude": 26.224525099062166,
        "longitude": 127.6807117301942,
        "radius": 50
    },
    "point_number_233": {
        "latitude": 26.224787716462288,
        "longitude": 127.68029914833569,
        "radius": 50
    },
    "point_number_234": {
        "latitude": 26.22500724177374,
        "longitude": 127.67985638395137,
        "radius": 50
    },
    "point_number_235": {
        "latitude": 26.225180914338356,
        "longitude": 127.67938900506027,
        "radius": 50
    },
    "point_number_236": {
        "latitude": 26.22530655012287,
        "longitude": 127.67890288922312,
        "radius": 50
    },
    "point_number_237": {
        "latitude": 26.225382569184397,
        "longitude": 127.67840414962863,
        "radius": 50
    },
    "point_number_238": {
        "latitude": 26.22540801553913,
        "longitude": 127.67789905821657,
        "radius": 50
    },
    "point_number_239": {
        "latitude": 26.225382569184397,
        "longitude": 127.6773939668045,
        "radius": 50
    },
    "point_number_240": {
        "latitude": 26.22530655012287,
        "longitude": 127.67689522721001,
        "radius": 50
    },
    "point_number_241": {
        "latitude": 26.225180914338356,
        "longitude": 127.67640911137286,
        "radius": 50
    },
    "point_number_242": {
        "latitude": 26.22500724177374,
        "longitude": 127.67594173248176,
        "radius": 50
    },
    "point_number_243": {
        "latitude": 26.224787716462288,
        "longitude": 127.67549896809744,
        "radius": 50
    },
    "point_number_244": {
        "latitude": 26.224525099062166,
        "longitude": 127.67508638623893,
        "radius": 50
    },
    "point_number_245": {
        "latitude": 26.224222692139556,
        "longitude": 127.6747091753625,
        "radius": 50
    },
    "point_number_246": {
        "latitude": 26.223884298636964,
        "longitude": 127.67437207911375,
        "radius": 50
    },
    "point_number_247": {
        "latitude": 26.223514174049033,
        "longitude": 127.67407933667363,
        "radius": 50
    },
    "point_number_248": {
        "latitude": 26.2231169729072,
        "longitude": 127.67383462944818,
        "radius": 50
    },
    "point_number_249": {
        "latitude": 26.222697690246292,
        "longitude": 127.67364103477271,
        "radius": 50
    },
    "point_number_250": {
        "latitude": 26.22226159878904,
        "longitude": 127.67350098721262,
        "radius": 50
    },
    "point_number_251": {
        "latitude": 26.221814182638546,
        "longitude": 127.67341624794722,
        "radius": 50
    },
    "point_number_252": {
        "latitude": 26.221361068312497,
        "longitude": 127.67338788262195,
        "radius": 50
    },
    "point_number_253": {
        "latitude": 26.220907953986448,
        "longitude": 127.67341624794722,
        "radius": 50
    },
    "point_number_254": {
        "latitude": 26.220460537835955,
        "longitude": 127.67350098721262,
        "radius": 50
    },
    "point_number_255": {
        "latitude": 26.2200244463787,
        "longitude": 127.67364103477271,
        "radius": 50
    },
    "point_number_256": {
        "latitude": 26.219605163717794,
        "longitude": 127.67383462944818,
        "radius": 50
    },
    "point_number_257": {
        "latitude": 26.21920796257596,
        "longitude": 127.67407933667363,
        "radius": 50
    },
    "point_number_258": {
        "latitude": 26.21883783798803,
        "longitude": 127.67437207911375,
        "radius": 50
    },
    "point_number_259": {
        "latitude": 26.21849944448544,
        "longitude": 127.6747091753625,
        "radius": 50
    },
    "point_number_260": {
        "latitude": 26.218197037562827,
        "longitude": 127.67508638623893,
        "radius": 50
    },
    "point_number_261": {
        "latitude": 26.217934420162706,
        "longitude": 127.67549896809744,
        "radius": 50
    },
    "point_number_262": {
        "latitude": 26.217714894851255,
        "longitude": 127.67594173248176,
        "radius": 50
    },
    "point_number_263": {
        "latitude": 26.217541222286638,
        "longitude": 127.67640911137286,
        "radius": 50
    },
    "point_number_264": {
        "latitude": 26.217415586502124,
        "longitude": 127.67689522721001,
        "radius": 50
    },
    "point_number_265": {
        "latitude": 26.217339567440597,
        "longitude": 127.6773939668045,
        "radius": 50
    },
    "point_number_266": {
        "latitude": 26.217314121085863,
        "longitude": 127.67789905821657,
        "radius": 50
    },
    "point_number_267": {
        "latitude": 26.217339567440597,
        "longitude": 127.67840414962863,
        "radius": 50
    },
    "point_number_268": {
        "latitude": 26.217415586502124,
        "longitude": 127.67890288922312,
        "radius": 50
    },
    "point_number_269": {
        "latitude": 26.217541222286638,
        "longitude": 127.67938900506027,
        "radius": 50
    },
    "point_number_270": {
        "latitude": 26.217714894851255,
        "longitude": 127.67985638395137,
        "radius": 50
    },
    "point_number_271": {
        "latitude": 26.217934420162706,
        "longitude": 127.68029914833569,
        "radius": 50
    },
    "point_number_272": {
        "latitude": 26.218197037562827,
        "longitude": 127.6807117301942,
        "radius": 50
    },
    "point_number_273": {
        "latitude": 26.21849944448544,
        "longitude": 127.68108894107064,
        "radius": 50
    },
    "point_number_274": {
        "latitude": 26.21883783798803,
        "longitude": 127.68142603731938,
        "radius": 50
    },
    "point_number_275": {
        "latitude": 26.21920796257596,
        "longitude": 127.6817187797595,
        "radius": 50
    },
    "point_number_276": {
        "latitude": 26.219605163717794,
        "longitude": 127.68196348698496,
        "radius": 50
    },
    "point_number_277": {
        "latitude": 26.2200244463787,
        "longitude": 127.68215708166042,
        "radius": 50
    },
    "point_number_278": {
        "latitude": 26.220460537835955,
        "longitude": 127.68229712922052,
        "radius": 50
    },
    "point_number_279": {
        "latitude": 26.220907953986448,
        "longitude": 127.68238186848592,
        "radius": 50
    },
    "point_number_280": {
        "latitude": 26.221361068312497,
        "longitude": 127.68291147554392,
        "radius": 50
    },
    "point_number_281": {
        "latitude": 26.221815982601488,
        "longitude": 127.68288575845159,
        "radius": 50
    },
    "point_number_282": {
        "latitude": 26.222266228854274,
        "longitude": 127.68280887106677,
        "radius": 50
    },
    "point_number_283": {
        "latitude": 26.222707186935022,
        "longitude": 127.68268160235806,
        "radius": 50
    },
    "point_number_284": {
        "latitude": 26.223134332017118,
        "longitude": 127.68250525827465,
        "radius": 50
    },
    "point_number_285": {
        "latitude": 26.223543281013992,
        "longitude": 127.68228164834545,
        "radius": 50
    },
    "point_number_286": {
        "latitude": 26.223929837555545,
        "longitude": 127.68201306711093,
        "radius": 50
    },
    "point_number_287": {
        "latitude": 26.22429003504856,
        "longitude": 127.68170227057803,
        "radius": 50
    },
    "point_number_288": {
        "latitude": 26.224620177379343,
        "longitude": 127.68135244793972,
        "radius": 50
    },
    "point_number_289": {
        "latitude": 26.22491687684082,
        "longitude": 127.68096718884966,
        "radius": 50
    },
    "point_number_290": {
        "latitude": 26.225177088895016,
        "longitude": 127.68055044658747,
        "radius": 50
    },
    "point_number_291": {
        "latitude": 26.225398143414136,
        "longitude": 127.6801064974927,
        "radius": 50
    },
    "point_number_292": {
        "latitude": 26.225577772079678,
        "longitude": 127.6796398970838,
        "radius": 50
    },
    "point_number_293": {
        "latitude": 26.225714131658457,
        "longitude": 127.6791554333124,
        "radius": 50
    },
    "point_number_294": {
        "latitude": 26.22580582291667,
        "longitude": 127.67865807743237,
        "radius": 50
    },
    "point_number_295": {
        "latitude": 26.22585190497793,
        "longitude": 127.67815293298807,
        "radius": 50
    },
    "point_number_296": {
        "latitude": 26.22585190497793,
        "longitude": 127.67764518344507,
        "radius": 50
    },
    "point_number_297": {
        "latitude": 26.22580582291667,
        "longitude": 127.67714003900076,
        "radius": 50
    },
    "point_number_298": {
        "latitude": 26.225714131658457,
        "longitude": 127.67664268312073,
        "radius": 50
    },
    "point_number_299": {
        "latitude": 26.225577772079678,
        "longitude": 127.67615821934933,
        "radius": 50
    },
    "point_number_300": {
        "latitude": 26.225398143414136,
        "longitude": 127.67569161894043,
        "radius": 50
    },
    "point_number_301": {
        "latitude": 26.225177088895016,
        "longitude": 127.67524766984566,
        "radius": 50
    },
    "point_number_302": {
        "latitude": 26.22491687684082,
        "longitude": 127.67483092758347,
        "radius": 50
    },
    "point_number_303": {
        "latitude": 26.224620177379343,
        "longitude": 127.67444566849342,
        "radius": 50
    },
    "point_number_304": {
        "latitude": 26.22429003504856,
        "longitude": 127.6740958458551,
        "radius": 50
    },
    "point_number_305": {
        "latitude": 26.223929837555545,
        "longitude": 127.6737850493222,
        "radius": 50
    },
    "point_number_306": {
        "latitude": 26.223543281013992,
        "longitude": 127.67351646808768,
        "radius": 50
    },
    "point_number_307": {
        "latitude": 26.223134332017118,
        "longitude": 127.67329285815849,
        "radius": 50
    },
    "point_number_308": {
        "latitude": 26.222707186935022,
        "longitude": 127.67311651407508,
        "radius": 50
    },
    "point_number_309": {
        "latitude": 26.222266228854274,
        "longitude": 127.67298924536637,
        "radius": 50
    },
    "point_number_310": {
        "latitude": 26.221815982601488,
        "longitude": 127.67291235798155,
        "radius": 50
    },
    "point_number_311": {
        "latitude": 26.221361068312497,
        "longitude": 127.67288664088922,
        "radius": 50
    },
    "point_number_312": {
        "latitude": 26.220906154023506,
        "longitude": 127.67291235798155,
        "radius": 50
    },
    "point_number_313": {
        "latitude": 26.22045590777072,
        "longitude": 127.67298924536637,
        "radius": 50
    },
    "point_number_314": {
        "latitude": 26.220014949689972,
        "longitude": 127.67311651407508,
        "radius": 50
    },
    "point_number_315": {
        "latitude": 26.219587804607876,
        "longitude": 127.67329285815849,
        "radius": 50
    },
    "point_number_316": {
        "latitude": 26.219178855611002,
        "longitude": 127.67351646808768,
        "radius": 50
    },
    "point_number_317": {
        "latitude": 26.21879229906945,
        "longitude": 127.6737850493222,
        "radius": 50
    },
    "point_number_318": {
        "latitude": 26.218432101576433,
        "longitude": 127.6740958458551,
        "radius": 50
    },
    "point_number_319": {
        "latitude": 26.21810195924565,
        "longitude": 127.67444566849342,
        "radius": 50
    },
    "point_number_320": {
        "latitude": 26.217805259784175,
        "longitude": 127.67483092758347,
        "radius": 50
    },
    "point_number_321": {
        "latitude": 26.217545047729978,
        "longitude": 127.67524766984566,
        "radius": 50
    },
    "point_number_322": {
        "latitude": 26.217323993210858,
        "longitude": 127.67569161894043,
        "radius": 50
    },
    "point_number_323": {
        "latitude": 26.217144364545316,
        "longitude": 127.67615821934933,
        "radius": 50
    },
    "point_number_324": {
        "latitude": 26.217008004966537,
        "longitude": 127.67664268312073,
        "radius": 50
    },
    "point_number_325": {
        "latitude": 26.216916313708325,
        "longitude": 127.67714003900076,
        "radius": 50
    },
    "point_number_326": {
        "latitude": 26.216870231647064,
        "longitude": 127.67764518344507,
        "radius": 50
    },
    "point_number_327": {
        "latitude": 26.216870231647064,
        "longitude": 127.67815293298807,
        "radius": 50
    },
    "point_number_328": {
        "latitude": 26.216916313708325,
        "longitude": 127.67865807743237,
        "radius": 50
    },
    "point_number_329": {
        "latitude": 26.217008004966537,
        "longitude": 127.6791554333124,
        "radius": 50
    },
    "point_number_330": {
        "latitude": 26.217144364545316,
        "longitude": 127.6796398970838,
        "radius": 50
    },
    "point_number_331": {
        "latitude": 26.217323993210858,
        "longitude": 127.6801064974927,
        "radius": 50
    },
    "point_number_332": {
        "latitude": 26.217545047729978,
        "longitude": 127.68055044658747,
        "radius": 50
    },
    "point_number_333": {
        "latitude": 26.217805259784175,
        "longitude": 127.68096718884966,
        "radius": 50
    },
    "point_number_334": {
        "latitude": 26.21810195924565,
        "longitude": 127.68135244793972,
        "radius": 50
    },
    "point_number_335": {
        "latitude": 26.218432101576433,
        "longitude": 127.68170227057803,
        "radius": 50
    },
    "point_number_336": {
        "latitude": 26.21879229906945,
        "longitude": 127.68201306711093,
        "radius": 50
    },
    "point_number_337": {
        "latitude": 26.219178855611002,
        "longitude": 127.68228164834545,
        "radius": 50
    },
    "point_number_338": {
        "latitude": 26.219587804607876,
        "longitude": 127.68250525827465,
        "radius": 50
    },
    "point_number_339": {
        "latitude": 26.220014949689972,
        "longitude": 127.68268160235806,
        "radius": 50
    },
    "point_number_340": {
        "latitude": 26.22045590777072,
        "longitude": 127.68280887106677,
        "radius": 50
    },
    "point_number_341": {
        "latitude": 26.220906154023506,
        "longitude": 127.68288575845159,
        "radius": 50
    }
}

def filter_place_types(types_list, allowed_types):
    return [t for t in types_list if t in allowed_types]

def estimate_price(price_level):
    if price_level == 1:
        return '〜1,000円'
    elif price_level == 2:
        return '1,000円〜3,000円'
    elif price_level == 3:
        return '3,000円〜5,000円'
    elif price_level == 4:
        return '5,000円以上'
    else:
        return '情報なし'

def translate_business_status(status):
    if status == "OPERATIONAL":
        return "営業中"
    elif status == "CLOSED_TEMPORARILY":
        return "一時閉店"
    elif status == "CLOSED_PERMANENTLY":
        return "恒久閉店"
    else:
        return "未設定"


def haversine(lat1, lon1, lat2, lon2):
    # 地球の半径 (メートル)
    R = 6371000

    # 緯度経度をラジアンに変換
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # 緯度と経度の差
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # ハーヴァーサイン式
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))

    # 距離 (メートル)
    distance = R * c
    return round(distance)

def format_opening_hours(opening_hours):
    if not opening_hours or 'weekday_text' not in opening_hours:
        return '情報なし'
    return '\n'.join(opening_hours['weekday_text'])


def fetch_data_for_points(api_key, place_types, fields, points):
    gmaps = googlemaps.Client(key=api_key)
    all_restaurants = []

    for point_name, point_info in points.items():
        latitude = point_info['latitude']
        longitude = point_info['longitude']
        radius = point_info['radius']
        location = (latitude, longitude)

        for place_type in place_types:
            place_count = 0  # 各ポイントで取得した場所の数をカウント

            # 初期リクエスト
            places_result = gmaps.places_nearby(location=location, radius=radius, type=place_type)
            all_restaurants.extend(process_places_results(places_result, gmaps, fields))
            place_count += len(places_result['results'])

            # 追加の結果がある場合は、next_page_tokenを使用して取得
            while 'next_page_token' in places_result:
                time.sleep(2)  # Googleのレートリミットに対応
                places_result = gmaps.places_nearby(page_token=places_result['next_page_token'])
                all_restaurants.extend(process_places_results(places_result, gmaps, fields))
                place_count += len(places_result['results'])

                if place_count >= 60:
                    print(f"警告: {point_name}の{place_type}で60件以上のデータを取得しています。")
                    break

    return all_restaurants

def process_places_results(places_result, gmaps, fields):
    results = []
    for place in places_result['results']:
        place_details = gmaps.place(place_id=place['place_id'], fields=fields)

        if 'result' in place_details:
            details = place_details['result']
            types_list = place.get('types', [])
            filtered_types = filter_place_types(types_list, place_types)

            # 基準点（RenSa）からの距離を計算
            distance = haversine(base_lat, base_lng, details['geometry']['location']['lat'], details['geometry']['location']['lng'])
            # print(details['geometry']['location']['lat'])
            # print(details['geometry']['location']['lng'])

            results.append({
                '店名': details.get('name', '情報なし'),
                '営業状況': translate_business_status(details.get('business_status', '情報なし')),
                'タイプ': ', '.join(filtered_types),
                '営業時間': format_opening_hours(details.get('opening_hours', {})),
                '価格帯推定': estimate_price(details.get('price_level', '情報なし')),
                'ユーザー評価': details.get('rating', '情報なし'),
                'ユーザー総評価数': details.get('user_ratings_total', '情報なし'),
                '距離（m）': distance,
                '住所': details.get('formatted_address', '情報なし'),
                '電話番号': details.get('formatted_phone_number', '情報なし'),
                'ウェブサイト': details.get('website', '情報なし'),
                'Google Map': f"https://www.google.com/maps/place/?q=place_id:{place['place_id']}"
            })
    return results


def upload_to_google_sheets(df, spreadsheet_name, worksheet_name, credentials_json):
    # GoogleスプレッドシートAPIの認証
    scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
    creds = ServiceAccountCredentials.from_json_keyfile_name(credentials_json, scope)
    client = gspread.authorize(creds)

    # スプレッドシートを開く（存在しない場合は新規作成）
    try:
        sheet = client.open(spreadsheet_name).worksheet(worksheet_name)
    except gspread.SpreadsheetNotFound:
        sheet = client.create(spreadsheet_name).sheet1
        sheet.update_title(worksheet_name)

    # DataFrameをスプレッドシートにアップロード
    sheet.update([df.columns.values.tolist()] + df.values.tolist())

# すべてのポイントからデータを取得
all_restaurants_data = fetch_data_for_points(api_key, place_types, fields, points)

# DataFrameへ変換し、重複を削除
all_restaurants_df = pd.DataFrame(all_restaurants_data).drop_duplicates(subset=['Google Map'])

# Googleスプレッドシートにアップロード
upload_to_google_sheets(all_restaurants_df, "google map survey for rensa", "シート1", "credentials.json")

print(f"{len(all_restaurants_df)}件取得完了")
```
